```python
#contamos valores NAN
#genero e salario
dataset.isnull().sum()
```

```python
#salarios
#remover NAs e substituir pela mediana
dataset["Salario"].describe()
```

```python
mediana = sts.median(dataset["Salario"])
mediana
```

```python
#substituir NAN por mediana
dataset.fillna({"Salario": mediana}, inplace=True)
```

```python
#verificamos se NAN n√£o existem mais
dataset['Salario'].isnull().sum()
```

```python
#genero
agrupado = dataset.groupby(["Genero"]).size()

agrupado
```

```python
#total de NAs
dataset['Genero'].isnull().sum()
```

```python
moda = sts.mode(dataset["Genero"])
moda
```

```python
#preenche NAs com masculino (moda)
dataset.fillna({"Genero": moda}, inplace=True)
```

```python
dataset['Genero'].isnull().sum()
```

```python
#padroniza de acordo com o dominio
dataset.loc[dataset['Genero'] == 'M', 'Genero'] = 'Masculino'
dataset.loc[dataset['Genero'].isin(['F', 'Fem']), 'Genero'] = 'Feminino'
#visualiza o resultado
agrupado = dataset.groupby(["Genero"]).size()
agrupado
```

```python
#idades fora do dominio
dataset['Idade'].describe()
```

```python
#visualizar
dataset.loc[(dataset['Idade'] < 0) | (dataset['Idade'] > 120)]
```

```python
#calcular a mediana
mediana = sts.median(dataset['Idade'])
mediana
```

```python
#substituir
dataset.loc[(dataset['Idade'] < 0) | (dataset['Idade'] > 120), 'Idade'] = mediana
```

```python
#verificamos se ainda existem idades fora do dominio
dataset.loc[(dataset['Idade'] < 0) | (dataset['Idade'] > 120)]
```

```python
#dados duplicados, buscamos pelo ID
dataset[dataset.duplicated(['ID'], keep=False)]
```

```python
#excluimos pelo ID
dataset.drop_duplicates(subset='ID', keep='first', inplace=True)
#buscamos duplicados
dataset[dataset.duplicated(['ID'], keep=False)]
```

```python
agrupado = dataset.groupby(["Estado"]).size()
agrupado
```

```python
#atribuimos RS (moda)
dataset.loc[dataset['Estado'].isin(['RP', 'SP', 'TD']), 'Estado'] = 'RS'
agrupado = dataset.groupby(["Estado"]).size()
agrupado
```

```python
#outliers em salario, vamos considerar 2 desvios padr√£o
desv = sts.stdev(dataset['Salario'])
desv
```

```python
#definir padr√£o como maior que 2 desvios padr√£o
#checamos se algum atende crit√©rio
dataset.loc[dataset['Salario'] >= 2 * desv]
```

```python
#vamos atualizar salarios para mediana, calculamos
mediana = sts.median(dataset['Salario'])
mediana
```

```python
#atribuimos
dataset.loc[dataset['Salario'] >= 2 * desv, 'Salario'] = mediana
#checamos se algum atende criterio
dataset.loc[dataset['Salario'] >= 2 * desv]
```
