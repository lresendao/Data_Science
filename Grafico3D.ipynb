```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Importa suporte para gr√°ficos 3D
import pandas as pd
```

```python
base = pd.read_csv('orchard.csv')
base.head()
```

```python
# Verifique se 'base' est√° corretamente definido
print(base.head())  # Apenas para depura√ß√£o

# Criando a figura e o eixo 3D
figura = plt.figure(figsize=(8, 6))  # Define o tamanho da figura
eixo = figura.add_subplot(111, projection='3d')  # Alternativa para (1,1,1)

# Criando o scatter plot 3D
eixo.scatter(base['decrease'], base['rowpos'], base['colpos'], c='r', marker='o')

# Configurando r√≥tulos dos eixos
eixo.set_xlabel('Decrease')
eixo.set_ylabel('Rowpos')
eixo.set_zlabel('Colpos')

plt.show()  # Exibe o gr√°fico
```
