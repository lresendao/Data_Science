```python
import pandas as pd
import seaborn as srn
import matplotlib as plt
```

```python
base = pd.read_csv('co2.csv')
base.head()
```

```python
#Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
srn.scatterplot(data=base, x="conc", y="uptake", hue="Type")
```

```python
#Seleção de registros específicos da base de dados (Quebec e Mississipi)
q = base.loc[(base['Type'] == 'Quebec')]
m = base.loc[(base['Type'] == 'Mississippi')]
q
m
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))  # Define o tamanho da figura

# Primeiro subplot - Quebec
ax1 = plt.subplot(1, 2, 1)
sns.scatterplot(data=q, x="conc", y="uptake", ax=ax1)
ax1.set_title('Quebec')

# Segundo subplot - Mississippi
ax2 = plt.subplot(1, 2, 2)
sns.scatterplot(data=m, x="conc", y="uptake", ax=ax2)
ax2.set_title('Mississippi')

plt.tight_layout()
plt.show()
```

```python
#Filtro de refrigerado e não refrigerado
ch = base.loc[(base['Treatment'] == 'chilled')]
nh = base.loc[(base['Treatment'] == 'nonchilled')]
```

```python
#Gráfico somente com 'chilled' e 'nonchilled'
plt.figure(figsize=(10, 5))  # Define o tamanho da figura

# Primeiro subplot - ch
plt.subplot(1, 2, 1)
sns.scatterplot(data=ch, x="conc", y="uptake").set_title('chilled')

# Segundo subplot - nh
plt.subplot(1, 2, 2)
sns.scatterplot(data=nh, x="conc", y="uptake").set_title('nonchilled')

plt.tight_layout()
```
